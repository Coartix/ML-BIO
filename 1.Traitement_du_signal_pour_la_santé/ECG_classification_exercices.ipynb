# Context
Cardiovascular Disease is the main cause of humain death, responsible for 31% of the world cide deaths in 2016. The major cause of the cardiac deaths is cardiac arrhythmia. The best equipment for monitoring and identifying arrhythmia is an electrocardiogram(ECG).

By recording ECG, analysing these features and classifying signals we can make assessments about the healthy state of the heart, diagnose certain diseases of the heart and predict whether a subject will go on to develop certain diseases.

ECG (Electrocardiogram) is non-stationary physiolocial signal representing electrical activity of heart. In other words, ECG is a periodic waveform that describes the action of heart as it moves.


In this project, we will classify ECG signals to make assessments about the health state of the heart. For this purpose, we use ensemble of three PhysioNet databases: MIT-BIH Arrhythmia Database, MIT-BIH Normal Sinus Rhythm Database, and The BIDMC Congestive Heart Failure Database.

- MIT-BIH Arrhythmia database (ARR-Database) consists of 48 half-hour ECG recording from people with cardiac arrhythmia problem. Each record was sampled at 360 hertz and contains two chanels. source [here](https://www.physionet.org/content/mitdb/1.0.0/ ).

- MIT-BIH Normal Sinus Rhythm Database (NSR-Database) consists of 18 long-tern ECG recording from people with normal sinus rhythms. Records in this databased were sampled at 128 hertz.
source [here](https://www.physionet.org/content/nsrdb/1.0.0/).

- BIDMC Congestive Heart Failure Database (CHF-Database) consists of 15 long-tern ECG recording from people with congestive heart failure. Records in this databased were sampled at 250 hertz. [source](https://www.physionet.org/content/chfdb/1.0.0/)


If you want to see waves : https://www.physionet.org/lightwave/?db=mitdb/1.0.0

```python
def load_record_using_wfdb(filename, start, end, channel=None, directory='mitdb'):
    return wfdb.rdsamp(filename, sampfrom=start, sampto=end, channels=channel, pn_dir=directory)
```

```python
start = 0
end = 2000
physical_signals = load_record_using_wfdb('100', start, end)

sampling_frequency = physical_signals[1]['fs']
time_values = [start + i / sampling_frequency for i in range(len(physical_signals[0]))]

plt.figure(figsize=(12, 6))
plt.plot(time_values, physical_signals[0][:, 0], label='Complexe QRS')
plt.plot(time_values, physical_signals[0][:, 1], label='Vague P')
plt.title(f'Signal from Record mitdb')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

Un motif identique se répète sur un intervalle de temps régulier, le signal semble être périodique. Essayons d'approximer la période de ce motif:

```python
autocorr_channel_0 = np.correlate(physical_signals[0][:, 0], physical_signals[0][:, 0], mode='full')
autocorr_channel_1 = np.correlate(physical_signals[0][:, 1], physical_signals[0][:, 1], mode='full')

# Plot the autocorrelation results
plt.figure(figsize=(12, 6))
plt.plot(autocorr_channel_0, label='Channel 0 Autocorrelation')
plt.plot(autocorr_channel_1, label='Channel 1 Autocorrelation')
plt.title('Autocorrelation of Signals')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Trouvons l'indice du premier pic significatif dans l'autocorrélation
threshold = 30/270
significant_peak_index = np.argmax(autocorr_channel_0 > threshold * max(autocorr_channel_0))

# Calculer la période en échantillons (lag)
period_in_samples = significant_peak_index

period_in_seconds = period_in_samples / sampling_frequency
print(f"La période est de environ {period_in_seconds} secondes")
```

Trouvons maintenant la corrélation avec une forme sinusoïdale depuis le channel 0:

```python
t = np.arange(len(physical_signals[0][:, 0]))
estimated_frequency = 1 / period_in_samples
reference_sine_wave = np.sin(2 * np.pi * estimated_frequency * t)

# Tracer le signal d'origine et la sinusoïde de référence
plt.figure(figsize=(12, 6))
plt.plot(t, physical_signals[0][:, 0], label='Signal channel_0', color='blue')
plt.plot(t, reference_sine_wave, label='Sinusoïde de Référence', linestyle='--', color='red')
plt.title('Comparaison entre le Signal et la Sinusoïde de Référence')
plt.xlabel('Échantillons')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
cross_correlation = np.correlate(physical_signals[0][:, 0], reference_sine_wave, mode='full')
plt.plot(cross_correlation)
```
